```python
import os
import datetime
now = datetime.datetime.now

from bokeh.io import output_notebook, show
from bokeh.palettes import Category10_3
output_notebook()

from data import bathtub
from plots import plot_x_y
from bokeh_saving import save_figures_button
from plot_utils import get_figure_size


PLOTS_DIR = os.path.abspath('plot_pdfs')
if not os.path.exists(PLOTS_DIR):
    os.mkdir(PLOTS_DIR)

FIG_SIZE = get_figure_size(ratio=1.3)
```

```python
xs, ys, xs_c, ys_c = bathtub()
```

```python
p = plot_x_y(None, xs, ys, xs_c, ys_c, *FIG_SIZE, colors=Category10_3, use_latex_style=True, title='')
show(p)

save_figures_button([('bathtub_fun', p)], svg_results_dir=PLOTS_DIR, png_results_dir=PLOTS_DIR)
```
