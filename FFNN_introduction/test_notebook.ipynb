```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(1,10,450)
ys = np.concatenate([np.square(xs[0:150]-4)+1, \
                        1+0.1*np.sin((xs[150:240]-3.9)*(np.pi/2)), np.ones(60), \
                        np.square(xs[300:450]-7)+1])

print(min(xs[0:150]), max(xs[0:150]))
print(min(xs[150:240]), max(xs[150:240]))
print(min(xs[240:300]), max(xs[240:300]))
print(min(xs[300:450]), max(xs[300:450]))
```

```python
plt.figure()
plt.plot(xs, ys)
```
